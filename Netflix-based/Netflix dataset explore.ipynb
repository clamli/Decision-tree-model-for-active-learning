```python
from scipy.sparse import *

rating_matrix_csc = load_npz('netflix/sparse_matrix_100%.npz').tocsc()
print("file load DONE")
```

```python
# item_lst = []
# for i in range(rating_matrix_csc.shape[0]):
#     if i % 100 == 0:
#         print("%.2f%%" % (100 * i / rating_matrix_csc.shape[0]))
#     item_lst.append(len(rating_matrix_csc[i,:].nonzero()[0]))
a = rating_matrix_csc[:,0:int(rating_matrix_csc.shape[1]*0.75)].getnnz(axis=1).tolist()
```

```python
usr_lst = []
for i in range(rating_matrix_csc.shape[1]):
    if i % 100 == 0:
        print("%.2f%%" % (100 * i / rating_matrix_csc.shape[1]))
    usr_lst.append(len(rating_matrix_csc[:,i].nonzero()[0]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

n, bins, patches = plt.hist(np.array(a[0:]), 1000, facecolor='g',cumulative=False, alpha=0.75)

plt.xlabel('number of ratings')
plt.ylabel('number of items')
plt.title('distribution of rating')
plt.axis([0, 30000, 0, 6000])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n, bins, patches = plt.hist(np.array(usr_lst[:]),1000, normed=1, facecolor='b',cumulative=True, alpha=0.75)

plt.xlabel('number of ratings')
plt.ylabel('number of users')
plt.title('distribution of rating')
plt.axis([0, 750, 0, 1])
plt.grid(True)
plt.show()
```
